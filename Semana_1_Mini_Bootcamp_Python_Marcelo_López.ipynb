# **Operaciones Básicas**

Si una operación involucra float, el resultado será float (excepto // que redondea hacia abajo)
La división / siempre devuelve float, incluso si es exacta (4 / 2 -> 2.0)
El operador % devuelve el resto de la división, también conocido como módulo.
El operador ** se usa para potencias. Por ejemplo: 2 ** 5 es 2^5

a = 10
b = 3

print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.333...
print(a // b)  # 3
print(a % b)   # 1
print(a ** b)  # 1000

# Divisiones en Python

División normal. Siempre devuelve un número decimal (float), aunque la división sea exacta.

División entera. Devuelve el resultado entero, eliminando la parte decimal. Redondea hacia abajo sin importar si el resultado es positivo o negativo.

Módulo. Devuelve el resto de la división.

# **f-strings**

```python
# f"Texto {variable}"
```

Esto permite insertar variables, expresiones y aplicar formato dentro de {}.

```python
nombre = "Ana"
edad = 25
print(f"Hola, {nombre}. Tienes {edad} años.")
# Hola, Ana. Tienes 25 años.
```

# **.format()**

Se pueden usar índices {0}, {1}, o nombres {nombre}.

```python
print("Hola, {}. Tienes {} años.".format("Ana", 25))
# Hola, Ana. Tienes 25 años.

print("Hola, {nombre}. Tienes {edad} años.".format(nombre="Ana", edad=25))
# Hola, Ana. Tienes 25 años.
```

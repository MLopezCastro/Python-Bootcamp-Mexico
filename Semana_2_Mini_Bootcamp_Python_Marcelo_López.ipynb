## Operadores Lógicos

# and (Y lógico)

Devuelve True solo si ambas condiciones son verdaderas.
Si alguna condición es falsa, el resultado es False.

```python
edad = 20
tiene_ine = True
print(edad >= 18 and tiene_ine)
```

# or (O lógico)

Devuelve True si al menos una condición es verdadera.
Solo devuelve False si todas son falsas.

```python
edad = 16
permiso_padres = True
print(edad >= 18 or permiso_padres)
```

# not (negación lógica)

not True se mueve a False

not False se mueve a True

```python
es_mayor = False
print(not es_mayor)
```

```python
# Ejemplo 1: Verificar acceso a un sistema
usuario = "Mike"
contraseña = "1234"
print(usuario == "Mike" and contraseña == "1234")
# True (ambas condiciones se cumplen)

# Ejemplo 2: Acceso con diferentes condiciones
tiene_boletos = False
esta_en_lista = True
print(tiene_boletos or esta_en_lista)
# True (aunque no tiene boletos, está en la lista)

# Ejemplo 3: Negar un valor
llueve = True
print(not llueve)
# False (porque sí está lloviendo)
```

# Prioridad y uso de paréntesis en operadores lógicos

En Python, los operadores lógicas tienen un orden de evaluación (de mayor a menor):

1º: not

2º: and

3º: or

Esto significa que si mezclas en una misma expresión, Python primero aplicará not, después and, y al final or.

```python
print(True or False and False)
# Paso 1: (False and False) → False
# Paso 2: True or False → True
# Resultado: True
```

## Para evitar confusiones y hacer más claro el código, se puede usar paréntesis:

```python
print((True or False) and False)
# (True or False) → True
# True and False → False
```

## Operaciones de corto circuito

Cuando el resultado ya se puede determinar con la primera parte de la expresión, Python se detiene y no sigue evaluando.

### `¿Cómo funciona?`

## AND (and)

Si el primer valor es False, la expresión completa será False, sin importar lo que venga después.

```python
print(False and 10/0)  # False (no intenta dividir entre 0)
```

## OR (or)

Si el primer valor es True, la expresión completa será True, sin evaluar lo demás.

```python
print(True or 10/0)  # True (no evalúa la división)
```

Regla clave:

Python no solo devuelve True o False, también devuelve el primer valor que determina el resultado.

Esto lo hace útil en condiciones y valores por defecto.

```python
print(0 or "Hola")    # "Hola" (0 es False → toma el segundo valor)
print("Python" and 5) # 5 (ambos True → devuelve el último)
print("" and 10)      # "" (cadena vacía es False → se detiene ahí)
```

Usos prácticos:

Evitar errores cuando hay valores nulos o vacíos

Esto es útil porque no intentamos procesar "No está vacío", si la primera parte ya es falsa.

```python
texto = ""
resultado = texto and "No está vacío"
print(resultado)  # "" (cadena vacía, se detiene ahí)
```

Seleccionar un valor por defecto con or:

Si nombre está vacío, automáticamente toma "Anónimo"

```python
nombre = ""
mostrar = nombre or "Anónimo"
print(mostrar)  # "Anónimo"
```

Encadenar evaluaciones de manera segura:

No llega a evaluar b porque a ya es falso.

```python
a = 0
b = 10
resultado = a and b
print(resultado)  # 0
```

Evaluar expresiones más cortas:

Si la primera parte fuera falsa, no seguiría evaluando.


```python
edad = 20
mensaje = (edad > 18) and "Mayor de edad"
print(mensaje)  # "Mayor de edad"
```

Atajos para condiciones encadenadas:

```python
x = 0
y = 5
print(x or y)   # 5
```

## Ejemplos prácticos

Después de toda la teoría, vamos a revisar ejemplos muy sencillos para que puedas compararlos próximos a los ejercicios y los puedas realizar.


```python
# 1. Strings vacíos como falsos
# "" se evalúa como False, mientras que cualquier string no vacío es True
print("" and "texto")      # ""   (and se detiene en el primero falso)
print("" or "texto")       # "texto"   (or sigue hasta encontrar un verdadero)
print("hola" and "mundo")  # "mundo"   (ambos verdaderos → devuelve el último)

# 2. Números: 0 es falso, cualquier otro número es verdadero
print(0 and 10)   # 0   (and se detiene en el primero falso)
print(0 or 10)    # 10  (or sigue hasta el primero verdadero)
print(5 and 10)   # 10  (ambos verdaderos → devuelve el último)
print(5 or 0)     # 5   (or se detiene en el primero verdadero)

# 3. Encadenamiento con or: busca el primer valor verdadero
print(0 or "" or [] or "Final")  # "Final"
# Todos los primeros valores son falsos (0, "", []),
# entonces devuelve el primero verdadero: "Final"

# 4. Comparaciones combinadas con short-circuit
edad = 20
print(edad > 18 and "Acceso permitido")  # "Acceso permitido"
print(edad < 18 or "Acceso permitido")   # "Acceso permitido"
# Aquí, la comparación devuelve True o False y se combina con otros valores.
```

## Cadena de Comparadores

Una cadena de comparadores te permite escribir varias comparaciones en una sola expresión.

```python
# Define variables with numerical values for comparison
a = 5
b = 10
c = 15

# Examples of chained comparisons
print(a < b < c)
print(a <= b <= c)

edad = 25
print(18 <= edad < 65)

x = 1
y = 2
z = 3
print(x != y != z)
```

Hay algunas equivalencias importantes.

```python
# Equivalencia 1
a < b < c
(a < b) and (b < c)

# Equivalencia 2
18 <= edad < 65
(18 <= edad) and (edad < 65)

# Equivalencia 3
a == b == c
(a == b) and (b == c)
```

Características que debes dominar:

Evalúa de izquierda a derecha y hace corto circuito.
Si la primera comparación ya es falsa, no evalúa al resto.

```python
3 < 2 < 100   # False (ni mira la segunda parte)
```

En lugar de escribir con and, puedes usar una sola línea

```python
edad = 30
print(18 <= edad < 65)   # True
```

Puedes mezclar comparadores estrictos y no estrictos

```python
a, b, c = 2, 5, 5
print(a < b <= c)  # True
```

Funciona con strings (comparación lexicográfica). Las mayúsculas y los acentos crean que los caracteres tengan un peso diferente más allá del abecedario.

```python
print("Ana" < "casa" < "cáscara")  # True
```

Tipos incompatibles fallan:


No puedes mezclar tipos.



```python
3 < "hola" # lanza TypeError
```

Ejemplos:


```python
# Edad dentro de un rango
edad = 25
print(18 <= edad < 65)  # True

# Igualdades encadenadas
print(7 == 7 == 7)      # True
print(7 == 8 == 7)      # False

# Strings
print("Z" < "a" < "c")  # True ('Z'<'a' y 'a'<'c')

# Con operaciones matemáticas
n = 12
print(2*3 < n <= 15)    # True (6<12 y 12<=15)
```

## Condicionales

Para las condicionales, existen unas reglas:

La condición debe evaluarse a un booleano (True o False)

La indentación (espacios) es obligatoria en el bloque

Puedes tener solo un if, if + else, y más if +uno o más elif + opcionalmente un else.

Características


El bloque if se ejecuta solo si la condición es verdadera.

elif significa "sino, si..." y se evalúa en orden.

else se ejecuta solo si ninguna condición previa fue verdadera.

Solo un bloque dentro de una cadena if/elif/else se ejecuta

Puedes anidar condicionales (if dentro de otro if)

Es importante dejar el espacio. Usa "TAB" en tu teclado para hacer la indentación.

### if simple

```python
edad = 20
if edad >= 18:
    print("Eres mayor de edad")
# Resultado: Eres mayor de edad
```

### if + else

```python
x = 5
if x % 2 == 0:
    print("Es par")
else:
    print("Es impar")
# Resultado: Es impar
```

### if + elif + else

```python
nota = 75
if nota >= 90:
    print("Excelente")
elif nota >= 70:
    print("Aprobado")
else:
    print("Reprobado")
# Resultado: Aprobado
```

### Condicional anidado

```python
numero = -3
if numero >= 0:
    if numero == 0:
        print("El número es cero")
    else:
        print("El número es positivo")
else:
    print("El número es negativo")
# Resultado: El número es negativo
```

Errores comunes y recomendaciones:

Ten cuidado de no confundir = con ==


Olvidar la indentación


Olvidar los dos puntos (:)


Escribir condiciones imposibles. Se puede volver "código muerto"

## Operadores ternarios

### Asignación rápida

```python
edad = 20
mensaje = "Mayor de edad" if edad >= 18 else "Menor de edad"
print(mensaje)
# Mayor de edad
```

```python
es_admin = True
print("Acceso permitido" if es_admin else "Acceso denegado")
# Acceso permitido
```

```python
nota = 85
resultado = "Excelente" if nota >= 90 else "Aprobado" if nota >= 60 else "Reprobado"
print(resultado)
# Aprobado
```

## for/while

En Python, los bucles (loops) permiten ejecutar un bloque de código varias veces sin tener que repetirlo manualmente.

## for

Se usa cuando sabes cuántas veces quieres repetir algo. Itera y no necesitas manejar un contador manual (el mismo for lo hace).

variable: toma cada valor del iterable (string, lista, rango, etc)

iterable: una secuencia de elementos que puede recorrerse

```python
for letra in "Hola":
    print(letra)

# H
# o
# l
# a
```

## while

Se usa cuando no sabes cuántas veces se repetirá el ciclo. Puede producir bucles infinitos si la condición nunca se vuelve False.

condicion: se evalúa en cada iteración.

El bucle continúa mientras la condición sea True.

```python
# Contar de 1 a 5
contador = 1
while contador <= 5:
    print(contador)
    contador += 1  # importante para que termine

# Pedir contraseña hasta que sea correcta
clave = ""
while clave != "python":
    clave = input("Introduce la clave: ")
print("¡Acceso permitido!")
```

Comparación práctica:

Usa for cuando quieres recorrer elementos de una colección o un rango de números.

Usa while cuando dependes de una condición lógica que puede cambiar en el tiempo.

Ejemplos

### Acumuladores en while

Muchas veces necesitas ir sumando valores dentro de un ciclo. Para eso usamos un acumulador.

```python
suma = 0
num = 1
while num <= 3:
    suma += num   # suma = suma + num
    num += 1
print("Suma final:", suma)

# Suma final: 6
```

### Validación con while y input

Es común pedir datos al usuario hasta que cumpla una condición. El usuario seguirá pidiendo la clave hasta que el usuario escriba 1234.

```python
clave = ""
while clave != "1234":
    clave = input("Introduce tu clave: ")
print("Correcto, acceso concedido.")
```

## **range()**

Cuando quieres repetir algo muchas veces con un for, necesitas una forma de decirle a Python "cuántas veces".

Ahí entra range().

La función range() se usa para generar secuencias de números, normalmente en combinación con los bucles for.

La función range() puede recibir hasta 3 valores.


inicio. es el número donde comienza. Si no lo escribes, Python asume que empieza en 0.


fin. es el número donde termina… pero ⚠️ no lo incluye. (Python se detiene justo antes).


paso. es cuánto aumenta o disminuye en cada vuelta. Si no lo escribes, Python asume 1.

### Solo el fin

```python
for i in range(5):
    print(i)
```

Genera 0, 1, 2, 3, 4

Empieza en 0 automáticamente

Se detiene en el 5, pero no lo incluye.

## Con inicio y fin

```python
for i in range(3, 7):
    print(i)
```

## break, continue, pass

Cuando usas for o while, no siempre quieres que el bucle recorra todo de principio a fin.

A veces necesitas interrumpir, saltar o simplemente dejar algo en blanco.

Para eso existen estas tres palabras clave.

## break

Rompe el bucle inmediatamente, sin importar si quedaban más interaciones.

En cuanto i == 5, se detiene todo el for.

```python
for i in range(10):
    if i == 5:
        break
    print(i)

# Salida
# 0
# 1
# 2
# 3
# 4
```

## continue

Salta la iteración actual y pasa a la siguiente.

Cuando i == 2, no ejecuta el print y sigue con 3.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)

# 0
# 1
# 3
# 4
```

## pass

Sirve como "relleno" cuando la sintaxis exige un bloque pero aún no tienes código.

En i == 1, no ocurrirá nada, pero el programa pasará y no fallará.

```python
for i in range(3):
    if i == 1:
        pass  # aquí no hacemos nada
    print(i)

# Salida
# 0
# 1
# 2
```

También es útil como marcador temporal en funciones o clases:

```python
def funcion_a_definir():
    pass
```

## for...else

En la mayoría de lenguajes de programación, los bucles for o while no llevan un else.

Pero en Python sí existe, y tiene un uso específico.

¿Cómo funciona?

La cláusula else en un bucle se ejecuta solo si el bucle terminó normalmente (es decir, sin un break).

Si el bucle se interrumpe con break, el else se salta.

## for con else (se ejecuta el else)

```python
for i in range(5):
    print(i)
else:
    print("El bucle terminó sin interrupciones")


# Salida:
# 0
# 1
# 2
# 3
# 4
# El bucle terminó sin interrupciones
```

## for con break (no se ejecuta el else)

```python
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("El bucle terminó sin interrupciones")


# Salida:
# 0
# 1
# 2
```

## Búsqueda en un string

El uso más práctico es cuando quieres buscar algo y saber si lo encontraste o no.


```python
texto = "python"
for letra in texto:
    if letra == "z":
        print("Encontré la letra z")
        break
else:
    print("No encontré la letra z")


# Salida:
# No encontré la letra z
```

Atenciones al concepto:

El else no funciona como un if. Aqui depende de break.

El else no se ejecuta siempre. Solo ocurre si el bucle termina normalmente.

## Precaución con ciclos infinitos

Un ciclo infinito es aquel que nunca termina porque la condición del bucle siempre es verdadera.

En la práctica, puede congelar tu programa o hacer que consuma demasiados recursos.

Ejemplo clásico con while:

Este ciclo nunca termina porque contador > 0 siempre será verdadero.

contador = 1

while contador > 0:

    print(contador)
    contador += 1

Ejemplo clásico con for:

En for es menos común, pero puede pasar si combinas mal range().

Esto no funcionará nunca.

for i in range(10, 0, 1):  # paso positivo pero inicio > fin

    print(i)

¿Cómo evitar estos problemas?:

Asegúrate de que la condición cambie

En while, incluye una instrucción que modifique la variable de la condición.

```python
x = 0
while x < 5:
    print(x)
    x += 1  # sin esto, nunca termina
```

Establece condiciones claras de salida:

Si usas input(), asegúrate de que haya una palabra clave para salir.

```python
clave = ""
while clave != "salir":
    clave = input("Escribe 'salir' para terminar: ")
```

Usa break como plan de escape:

```python
while True:  # bucle infinito
    dato = input("Escribe 'stop' para detener: ")
    if dato == "stop":
        break
```

## Nested Loops

Un ciclo anidado es simplemente un bucle dentro de otro bucle.

El bucle externo controla cuántas veces se repite el interno.

Sintaxis:

El for interno se completa todas sus vueltas antes de que el externo pase a la siguiente iteración.

```python
for i in range(3):          # bucle externo
    for j in range(2):      # bucle interno
        print(i, j)

# 0 0
# 0 1
# 1 0
# 1 1
# 2 0
# 2 1
```

Usos prácticos:

Recorrer tablas (filas y columnas)

Comparar elementos de una secuencia contra otra

Procesar strings, caracter por caracter, en múltiples niveles.

## doble for

```python
for fila in range(1, 4):       # 1,2,3
    for columna in range(1, 3): # 1,2
        print(f"Fila {fila}, Columna {columna}")


# Fila 1, Columna 1
# Fila 1, Columna 2
# Fila 2, Columna 1
# Fila 2, Columna 2
# Fila 3, Columna 1
# Fila 3, Columna 2
```

## patrón con strings

```python
for i in range(3):
    for j in range(3):
        print("*", end=" ")
    print()
```

## bucle while dentro de for

```python
for i in range(3):
    j = 0
    while j < 2:
        print(i, j)
        j += 1
```

Consideraciones importantes:

Los ciclos anidados crecen en complejidad muy rápido:

Un for normal. Significa "n" repeticiones

Dos for anidados. Significa "n" x "m" repeticiones

Evita anidar más de dos niveles salvo que sea necesario.
